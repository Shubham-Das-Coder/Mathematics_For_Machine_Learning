```python
#Addition of vectors
vector_1=np.asarray([0,0,2,3])
vector_2=np.asarray([0,0,3,-2])
sum=vector_1+vector_2
print(sum)
```

```python
plt.quiver(0,0,2,3,scale_units='xy',angles='xy',scale=1,color='b')
plt.quiver(0,0,3,-2,scale_units='xy',angles='xy',scale=1,color='y')
plt.quiver(0,0,5,1,scale_units='xy',angles='xy',scale=1,color='r')
plt.xlim(-8,8)
plt.ylim(-8,8)
plt.show()
```

```python
#Subtraction of vectors
vector_1=np.asarray([0,0,2,3])
vector_2=np.asarray([0,0,3,-2])
difference=vector_1-vector_2
print(difference)
```

```python
plt.quiver(0,0,2,3,scale_units='xy',angles='xy',scale=1,color='b')
plt.quiver(0,0,3,-2,scale_units='xy',angles='xy',scale=1,color='y')
plt.quiver(0,0,-1,5,scale_units='xy',angles='xy',scale=1,color='r')
plt.xlim(-8,8)
plt.ylim(-8,8)
plt.show()
```

```python
#Dot product of two vectors
a=np.array([2,3])
b=np.array([4,4])
a_dot_b=np.dot(a,b)
print(a_dot_b)
```

```python
#Dot product of two vectors
c=np.array([4,5,7])
d=np.array([2,8,3])
c_dot_d=np.dot(c,d)
print(c_dot_d)
```

```python
#Cross product of two vectors
a=np.array([2,3])
b=np.array([4,4])
a_cross_b=np.cross(a,b)
print(a_dot_b)
```

```python
#Cross product of two vectors
c=np.array([5,10,20])
d=np.array([18,32,50])
c_cross_d=np.cross(c,d)
print(c_cross_d)
```

```python
#Projection of a vector on another vector
x=np.array([2,5])
y=np.array([8,-6])

magnitude_of_y=np.sqrt((y[0]**2)+(y[1]**2))

projection_of_x_on_y=(np.dot(x,y)/magnitude_of_y**2)*y

print("Projection of x on y = ",projection_of_x_on_y)
```

```python
#Projection of a vector on another vector
x=np.array([3,4,5])
y=np.array([6,8,10])

magnitude_of_y=np.sqrt((y[0]**2)+(y[1]**2)+(y[2]**2))

projection_of_x_on_y=(np.dot(x,y)/magnitude_of_y**2)*y

print("Projection of x on y = ",projection_of_x_on_y)
```
